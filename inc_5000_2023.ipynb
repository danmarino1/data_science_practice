```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import json

import time
import random
```

```python
r = requests.get('https://api.inc.com/rest/i5list/2023').json()
#Save the JSON data to a file
with open('inc_500_all_data_2023.json', 'w') as f:
    json.dump(r, f)
```

```python
data = pd.DataFrame(r['companies'])
data.to_parquet('inc_5000_2023.parquet', index=False)
data
```

```python
r = requests.get('https://api.inc.com/rest/i5list/2022').json()
#Save the JSON data to a file
with open('inc_500_all_data_2022.json', 'w') as f:
    json.dump(r, f)

data_2022 = pd.DataFrame(r['companies'])
data_2022.to_parquet('inc_5000_2022.parquet', index=False)
data_2022
```
